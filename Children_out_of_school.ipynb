```python
# get the sum of student out of school in the world in each year

def sum_of_student(row_of_data):
    sum = 0
    for i in range(len(row_of_data)):
        if pd.isna(row_of_data[i]):
            sum += 0
        else:
            sum += row_of_data[i]
    return sum

Sum = []
for i in range (4, len(df.loc[0])-1): #4 is the first year
    Sum.append(sum_of_student(df.iloc[:,i]))

print(Sum)
```

```python


count_row = df.shape[0]

percentage = []

def percentage_of_student(order_in_sum_list):
    year = order_in_sum_list + 4
    percentage = []
    for k in range (0, count_row):
        if pd.isna(df.iloc[:,year][k]) or Sum[order_in_sum_list] == 0:
            percentage.append(0)
        else:
            percentage.append(df.iloc[:,year][k]/Sum[order_in_sum_list])
    return percentage
        
df2 = pd.DataFrame()
df2.insert (0, "Country Name", df.iloc[:,0])

for i in range (0, len(Sum)-1): 
    print(len(percentage_of_student(i)))
    df2.insert (i+1, df.columns[i+4], percentage_of_student(i))

df2.to_excel('/Users/walterwei/Desktop/Code/School/EM202/Leson 1/output.xlsx', index = False)

df2_transposed = df2.set_index('Country Name').T

df2_transposed.to_excel('/Users/walterwei/Desktop/Code/School/EM202/Leson 1/output_transposed.xlsx', index = True)
```
